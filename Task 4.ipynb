```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))
sns.barplot(x="Species", y="PetalWidthCm", data=iris, ax=axes[0,0], palette="rocket")
sns.barplot(x="Species", y="PetalLengthCm", data=iris, ax=axes[0,1], palette="mako")
sns.barplot(x="Species", y="SepalLengthCm", data=iris,ax=axes[1,0], palette="magma")
sns.barplot(x="Species", y="SepalWidthCm", data=iris,ax=axes[1,1], palette="viridis")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))
sns.violinplot(x="Species", y="PetalWidthCm", data=iris, ax=axes[0,0], palette="ch:start=5,rot=0", as_cmap=True)
sns.violinplot(x="Species", y="PetalLengthCm", data=iris, ax=axes[0,1], palette="ch:start=4,rot=0",as_cmap=True)
sns.violinplot(x="Species", y="SepalLengthCm", data=iris,ax=axes[1,0], palette="ch:start=3,rot=0",as_cmap=True)
sns.violinplot(x="Species", y="SepalWidthCm", data=iris,ax=axes[1,1], palette="ch:start=2,rot=0",as_cmap=True)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,12))
sns.boxplot(x="Species", y="PetalWidthCm", data=iris, ax=axes[0,0],palette="Blues")
sns.boxplot(x="Species", y="PetalLengthCm", data=iris, ax=axes[0,1],palette="Reds_r" )
sns.boxplot(x="Species", y="SepalLengthCm", data=iris,ax=axes[1,0], palette="Greens_r")
sns.boxplot(x="Species", y="SepalWidthCm", data=iris,ax=axes[1,1],palette="Purples_r")
plt.show()
```
